#  รู้จักกับภาษา C#




## ประวัติความเป็นมาและคุณสมบัติของภาษา C#

### ประวัติความเป็นมา

C# ถูกพัฒนาขึ้นในช่วงปลายยุค 90s โดยทีมงานที่ Microsoft นำโดย Anders Hejlsberg ซึ่งเป็นนักวิทยาการคอมพิวเตอร์ชาวเดนมาร์กที่มีชื่อเสียง  เขาเป็นผู้อยู่เบื้องหลังภาษาโปรแกรมที่สำคัญหลาย ๆ ภาษาก่อนหน้า C# เขาเป็นที่รู้จักจากผลงานการพัฒนา Turbo Pascal และ Delphi ให้กับบริษัท Borland

C#  ถูกออกแบบมาให้เป็นภาษาโปรแกรมเชิงวัตถุ (Object-Oriented Programming) ที่ทันสมัย ทำงานบน .NET Framework  จุดมุ่งหมายคือต้องการสร้างภาษาที่ใช้งานง่าย  มีประสิทธิภาพ  และสามารถใช้พัฒนาแอพพลิเคชั่นได้หลากหลายประเภท

C# เป็นภาษาที่ได้รับความนิยมอย่างต่อเนื่อง และมีการพัฒนาอย่างต่อเนื่องเช่นกัน ปัจจุบัน C# อยู่ในเวอร์ชัน 11

#### การพัฒนาของภาษา C#

C# เริ่มพัฒนาในปี 1999 และเปิดตัวครั้งแรกในปี 2000 พร้อมกับ .NET Framework โดยในช่วงแรก C# มีชื่อเรียกว่า "Cool" ซึ่งย่อมาจาก "C-like Object Oriented Language"
แต่ต่อมาได้เปลี่ยนชื่อเป็น C# ที่เลียนแบบภาษา C++ 

เครื่องหมาย ++ ในภาษาซี คือการเพิ่มค่าขึ้น 1 ระดับ (incremental) ดังนั้นเครื่องหมาย # จึงเทียบได้กับเครื่องหมาย ++ ซ้อนกันอยู่จำนวน 2 ชั้นและอาจแปลได้ว่าเพิ่มขึ้นในทุกมิตินั่นเอง


#### แรงบันดาลใจในการสร้างภาษา  C#

ภาษา C# ได้รับอิทธิพลจากภาษาโปรแกรมหลายภาษา เช่น C++ Java และ Delphi

##### อิทธิพลจากภาษา Java

__Garbage Collection__  C# ใช้ระบบ garbage collection คล้ายกับ Java ซึ่งช่วยจัดการหน่วยความจำโดยอัตโนมัติ นักพัฒนาไม่ต้องกังวลเรื่องการจัดสรรและคืนหน่วยความจำเอง ลดโอกาสเกิดข้อผิดพลาด และทำให้การพัฒนาโปรแกรมง่ายขึ้น

__Platform Independence__  แม้ว่าในช่วงแรก C# จะเน้นการทำงานบนแพลตฟอร์ม Windows แต่แนวคิดเรื่อง platform independence จาก Java ก็มีอิทธิพลต่อการพัฒนา .NET Core ในปัจจุบัน ซึ่งทำให้ C# สามารถทำงานบนแพลตฟอร์มอื่นๆ ได้เช่นกัน


##### อิทธิพลจากภาษา C++

__ประสิทธิภาพ__  C# ถูกออกแบบมาให้มีประสิทธิภาพสูง สามารถทำงานได้รวดเร็ว ใกล้เคียงกับ C++ ซึ่งเป็นภาษาที่ขึ้นชื่อเรื่องประสิทธิภาพ

__ความสามารถในการควบคุมระดับล่าง__ C# อนุญาตให้นักพัฒนาสามารถเข้าถึงและควบคุมฮาร์ดแวร์ หน่วยความจำ และทรัพยากรระบบในระดับล่างได้ คล้ายกับ C++ ซึ่งทำให้สามารถพัฒนาแอพพลิเคชันที่ต้องการประสิทธิภาพสูง เช่น เกม หรือโปรแกรมควบคุมอุปกรณ์


##### อิทธิพลจากภาษา  Delphi:

__ความง่ายในการใช้งาน__  C# ได้รับอิทธิพลจาก Delphi ในเรื่องของความง่ายในการใช้งาน มี syntax ที่เข้าใจง่าย และเครื่องมือพัฒนาที่ใช้งานสะดวก ช่วยลดความซับซ้อนในการพัฒนาโปรแกรม

__Rapid Application Development (RAD)__  C# และ .NET Framework สนับสนุนการพัฒนาแอพพลิเคชันแบบ RAD เช่นเดียวกับ Delphi โดยมีเครื่องมือและไลบรารีที่ช่วยให้สามารถสร้างแอพพลิเคชันได้อย่างรวดเร็ว

#### มาตรฐาน

C# ได้รับการรับรองเป็นมาตรฐานโดย ECMA International (ECMA-334) และ ISO/IEC (ISO/IEC 23270)

C#  ได้รับความนิยมอย่างต่อเนื่อง  มีการพัฒนาและปรับปรุงเวอร์ชันใหม่ๆ ออกมาเรื่อยๆ  ปัจจุบันเป็นหนึ่งในภาษาโปรแกรมหลักที่ใช้ในการพัฒนาแอพพลิเคชันบนแพลตฟอร์มของ Microsoft  และยังสามารถใช้พัฒนาแอพพลิเคชันบนแพลตฟอร์มอื่นๆ ได้ด้วย  เช่น  Linux  macOS  และ Android ผ่าน .NET Core
